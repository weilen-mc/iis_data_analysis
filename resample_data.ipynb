```python
def load_and_modify_hdf(input_hdf_path, key, n):
   
    # Load the dataset from the original HDF file
    with pd.HDFStore(input_hdf_path, mode='r') as store:
        if key not in store:
            raise KeyError(f"The key '{key}' is not in the HDF file.")
        
        # Load the specified dataset
        df = store[key]
    
    # Select every nth row from the DataFrame
    df_modified = df.iloc[::n, :]
    
    # Write the modified DataFrame and other groups into a new HDF file
    with pd.HDFStore(input_hdf_path, mode='w') as store_out:
        # Write the modified DataFrame
        store_out.put(key, df_modified)
        
```

```python
def load_and_modify_hdf(input_hdf_path, key, n):
    # Open the original HDF file
    with pd.HDFStore(input_hdf_path, mode='r+') as store:
        if key not in store:
            raise KeyError(f"The key '{key}' is not in the HDF file.")
        
        # Load the specified dataset
        df = store[key]
        
        # Select every nth row from the DataFrame
        df_modified = df.iloc[::n, :]

        store.put(key, df_modified)
```

```python
interleave = 50

for f_k, f_v in all_files.items():
    filename = f_v['filename']
    path = directory / Path(filename)
    out_filename = filename[:-3] + f'_subsampled_{interleave}.h5'
    out_path = directory / Path(out_filename)
    load_and_modify_hdf(path, 'results', interleave)
    
```
